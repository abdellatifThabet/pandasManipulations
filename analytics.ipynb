```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("result_svs.csv")
```

```python
df.columns
```

```python
df = df.melt(id_vars=["id","name","description","image"], 
        var_name="trait_type", 
        value_name="value")
```

```python
df = df[["id","trait_type","value"]]
```

```python
total = df['id'].nunique()
```

```python
df.head()
```

```python
count_list = df.groupby(["trait_type", "value"]).count()
```

```python
count_list.reset_index(inplace=True)  
```

```python
count_list
```

```python
count_list.rename(columns={'id': 'count'}, inplace=True)
```

```python

#count_list['value'] = count_list.index
#count_list.reset_index(drop=True, inplace=True)
#count_list = count_list[['trait_type','value']]
```

```python
count_list['rarity'] = np.nan
count_list['weight'] = 1
```

```python
for index in range(0, len(count_list)):
    if(str(count_list.iloc[index]['value']) == 'nan'):
        continue
    else:
        count_list.at[index,'rarity'] = (total/count_list.iloc[index]['count'])*count_list.iloc[index]['weight']

    
count_list['rarity'] = count_list['rarity'].astype('float64').round(2) 
```

```python
count_list
```

```python
count_list.to_csv('rarity.csv')
```

```python
dfo = pd.read_csv('result_svs.csv', index_col=[0])
```

```python
dfo.head()
```

```python
dfr = pd.read_csv('rarity_result_svs.csv', index_col=[0])
```

```python
dfr.head()
```

```python
dfo.head()
```

```python
index = 0
for row in dfo.iterrows():
    score = 0
    for col in range(3,len(dfo.columns)-1):
        if(str(row[1][col]) == 'nan'):
            continue
        else:
            col_name = dfo.columns[col]
            col_value = row[1][col]
            score = score + float(dfr.loc[(dfr.trait_type == col_name)&(dfr.value == col_value), 'rarity'])
    dfo.loc[row[0],'total_score'] = score
    index = index+1
```
